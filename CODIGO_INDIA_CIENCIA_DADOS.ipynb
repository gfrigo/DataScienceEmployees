# ***6) Há uma diminuição no número de funcionários à medida que os anos de experiência aumentam? (técnica de groupby)***
(Luiz Henrique de Albuquerque , Gabriel Frigo Sena Silva, Juan Carlos Hernandez)

```python
# Contagem dos anos de experiência
df_qntExperiencia = df['ExperienceInCurrentDomain'].value_counts().reset_index()
df_qntExperiencia.columns = ['ExperienceInCurrentDomain', 'Quantidade de Funcionários']

# Ordenar o DataFrame por anos de experiência em ordem crescente
df_qntExperiencia = df_qntExperiencia.sort_values(by='ExperienceInCurrentDomain')


# Criar um gráfico de colunas
fig_experiencia = px.bar(df_qntExperiencia, x='ExperienceInCurrentDomain', y='Quantidade de Funcionários',
                               labels={'Quantidade de Funcionários': 'Quantidade'},
                               title='Gráfico de Colunas de Anos de Experiência em Ordem Crescente')
fig_experiencia.show()


# Contagem dos anos de experiência
df_qntExperiencia = df['ExperienceInCurrentDomain'].value_counts().reset_index()
df_qntExperiencia.columns = ['ExperienceInCurrentDomain', 'Quantidade de Funcionários']

# Ordenar o DataFrame por anos de experiência em ordem crescente
df_qntExperiencia = df_qntExperiencia.sort_values(by='ExperienceInCurrentDomain')

# Encontre a maior torre
maior_torre = df_qntExperiencia[df_qntExperiencia['Quantidade de Funcionários'] == df_qntExperiencia['Quantidade de Funcionários'].max()]

# Encontre a menor torre
menor_torre = df_qntExperiencia[df_qntExperiencia['Quantidade de Funcionários'] == df_qntExperiencia['Quantidade de Funcionários'].min()]

# Faça um somatório de todas as torres
total_pessoas = df_qntExperiencia['Quantidade de Funcionários'].sum()

print("Maior torre:", maior_torre['ExperienceInCurrentDomain'].values[0], "com", maior_torre['Quantidade de Funcionários'].values[0], "pessoas")
print("Menor torre:", menor_torre['ExperienceInCurrentDomain'].values[0], "com", menor_torre['Quantidade de Funcionários'].values[0], "pessoas")
print("Total de pessoas em todas as torres:", total_pessoas)
```
